# Visual Data Exploration with Matplotlib

```python
data1 = np.random.normal(0, 0.1, 1000)
data2 = np.random.normal(1, 0.4, 1000) + np.linspace(0, 1, 1000)
data3 = 2 + np.random.random(1000) * np.linspace(1, 5, 1000)
data4 = np.random.normal(3, 0.2, 1000) + 0.3 * np.sin(np.linspace(0, 20, 1000))
```

```python
data = np.vstack([data1, data2, data3, data4]).transpose()
```

```python
df = pd.DataFrame(data, columns=['data1', 'data2', 'data3', 'data4'])
df.head()
```

### Line Plot

```python
df.plot(title='Line plot')
```

```python
plt.plot(df)
plt.title('Line plot')
plt.legend(['data1', 'data2', 'data3', 'data4'])
```

### Scatter Plot

```python
df.plot(style='.')
```

```python
_ = df.plot(kind='scatter', x='data1', y='data2',
            xlim=(-1.5, 1.5), ylim=(0, 3))
```

### Histograms

```python
df.plot(kind='hist',
        bins=50,
        title='Histogram',
        alpha=0.6)
```

### Cumulative distribution

```python
df.plot(kind='hist',
        bins=100,
        title='Cumulative distributions',
        density=True,
        cumulative=True,
        alpha=0.4)
```

### Box Plot

```python
df.plot(kind='box',
        title='Boxplot')
```

### Subplots

```python
fig, ax = plt.subplots(2, 2, figsize=(6, 6))

df.plot(ax=ax[0][0],
        title='Line plot')

df.plot(ax=ax[0][1],
        style='o',
        title='Scatter plot')

df.plot(ax=ax[1][0],
        kind='hist',
        bins=50,
        title='Histogram')

df.plot(ax=ax[1][1],
        kind='box',
        title='Boxplot')

plt.tight_layout()
```

### Pie charts

```python
gt01 = df['data1'] > 0.1
piecounts = gt01.value_counts()
piecounts
```

```python
piecounts.plot(kind='pie',
               figsize=(5, 5),
               explode=[0, 0.15],
               labels=['<= 0.1', '> 0.1'],
               autopct='%1.1f%%',
               shadow=True,
               startangle=90,
               fontsize=16)
```

### Hexbin plot

```python
data = np.vstack([np.random.normal((0, 0), 2, size=(1000, 2)),
                  np.random.normal((9, 9), 3, size=(2000, 2))])
df = pd.DataFrame(data, columns=['x', 'y'])
```

```python
data
```

## Exercise 2

```python
df = pd.read_csv('../data/weight-height.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Gender'].value_counts()
```

```python
males = df[df['Gender'] == 'Male']
females = df.query('Gender == "Female"')
fig, ax = plt.subplots()
males.plot(kind='scatter',
        x='Height',
        y='Weight',
        title='Ex2 graph',
        color='blue',
        ax=ax,
        alpha=0.3,
        )
females.plot(kind='scatter',
        x='Height',
        y='Weight',
        ax=ax,
        color='red',
        alpha=0.3,
        )
```

```python
df['GenderColor'] = df['Gender'].map({'Male': 'blue', 'Female': 'red'})
df.head()
```

```python
df.plot(kind='scatter',
        x='Height',
        y='Weight',
        title='Ex2 graph',
        c=df['GenderColor'],
        alpha=0.3,
        )
```

```python
fig, ax = plt.subplots()
ax.plot(males['Height'], males['Weight'], 'ob',
        females['Height'], females['Weight'], 'or', alpha=0.3)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Ex2 graph')
```

## Exercise 3
- plot the histogram of the heights for males and for females on the same plot
- use alpha to control transparency in the plot comand
- plot a vertical line at the mean of each population using `plt.axvline()`

```python
males['Height'].plot(kind='hist', bins=50, alpha=0.3, color='blue')
females['Height'].plot(kind='hist', bins=50, alpha=0.3, color='red')

plt.title('Height distribuion')
plt.legend(['Males', 'Females'])
plt.xlabel('Height (in)')

plt.axvline(x=males['Height'].mean(), color='blue', linewidth=2, linestyle='--', alpha=0.3)
plt.axvline(x=females['Height'].mean(), color='red', linewidth=2, linestyle='--', alpha=0.3)
```
